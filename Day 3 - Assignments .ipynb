# 2. Use For Loop to Print Prime Numbers in between 1 to 1000

```python
a=1
b=1001
for item in range(a,b):
    for i in range(2,item): 
        if((item%i)==0): 
            break 
    else: 
        print("Prime No's:",item) 
```
